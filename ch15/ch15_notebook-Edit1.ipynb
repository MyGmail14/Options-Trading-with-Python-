## Introduction to Portfolio Optimization and Risk Management

```python
# Suppose we have two AAPL options with the following expected returns and standard deviations:
# Option1: Expected Return = 0.08, Standard Deviation = 0.2
# Option2: Expected Return = 0.12, Standard Deviation = 0.3

# Portfolio weights (50% in each option)
weights = np.array([0.5, 0.5])

# Expected returns and standard deviations
returns = np.array([0.08, 0.12])
std_devs = np.array([0.2, 0.3])

# Calculate portfolio expected return and standard deviation
port_return = np.dot(weights, returns)
port_std_dev = np.sqrt(np.dot(weights**2, std_devs**2))

print(f"Portfolio Expected Return: {port_return:.2f}")
print(f"Portfolio Standard Deviation: {port_std_dev:.2f}")
```

```python
# Sample returns and covariance matrix for two AAPL options
returns = np.array([0.08, 0.12])
cov_matrix = np.array([[0.04, 0.02], [0.02, 0.09]])  # Assuming some covariance between the options

def negative_sharpe(weights): 
    port_return = np.dot(weights, returns)
    port_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    return -port_return / port_volatility

# Constraints and bounds
cons = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1}) 
bounds = ((0, 1), (0, 1))
initial_weights = [0.5, 0.5]

solution = minimize(negative_sharpe, initial_weights, method='SLSQP', bounds=bounds, constraints=cons)
optimal_weights = solution.x

print(f"Optimal Weights: {optimal_weights}")
```

## Diversification and Modern Portfolio Theory in Options Trading

```python
# Assuming a correlation of -0.5 between the two options
correlation = -0.5
covariance = correlation * std_devs[0] * std_devs[1]

# Portfolio variance formula considering covariance
port_variance = weights[0]**2 * std_devs[0]**2 + weights[1]**2 * std_devs[1]**2 + 2 * weights[0] * weights[1] * covariance
port_std_dev_diversified = np.sqrt(port_variance)

print(f"Portfolio Standard Deviation with Diversification: {port_std_dev_diversified:.2f}")
```

```python
num_portfolios = 10000
results = np.zeros((3, num_portfolios))
for i in range(num_portfolios):
    weights = np.random.random(2)
    weights /= np.sum(weights)
    
    port_return = np.dot(weights, returns)
    port_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    results[0,i] = port_return
    results[1,i] = port_volatility
    results[2,i] = results[0,i] / results[1,i]  # Sharpe Ratio

# Plotting
plt.scatter(results[1,:], results[0,:], c=results[2,:], cmap='YlGnBu', marker='o')
plt.title('Efficient Frontier with AAPL Options')
plt.xlabel('Portfolio Volatility')
plt.ylabel('Portfolio Return')
plt.colorbar(label='Sharpe Ratio')
plt.show()
```

## Options Strategies and Risk Management Techniques in Options Trading

### Implementing Simple Protective Measures with Options

#### Protective Puts

```python
# Parameters
stock_price = 100  # Current stock price
strike_price = 95  # Put option strike price
premium_put = 3    # Cost of the put option

def protective_put_payoff(s):
    return max(strike_price - s, 0) - premium_put + (s - stock_price)

stock_prices = np.arange(80, 120, 1)
payoffs = [protective_put_payoff(s) for s in stock_prices]

plt.plot(stock_prices, payoffs)
plt.title('Protective Put Payoff')
plt.xlabel('Stock Price at Expiry')
plt.ylabel('Profit/Loss')
plt.grid(True)
plt.show()
```

#### Covered Calls

```python
# Parameters for Covered Call
strike_price_call = 110  # Call option strike price
premium_call = 5         # Income received from selling the call

def covered_call_payoff(s):
    if s > strike_price_call:
        return stock_price - s + strike_price_call + premium_call - stock_price
    else:
        return premium_call

payoffs = [covered_call_payoff(s) for s in stock_prices]

plt.plot(stock_prices, payoffs)
plt.title('Covered Call Payoff')
plt.xlabel('Stock Price at Expiry')
plt.ylabel('Profit/Loss')
plt.grid(True)
plt.show()
```

#### Collars

```python
# Parameters for Collar
premium_put = 4
premium_call = 4
strike_price_put = 95
strike_price_call = 110

def collar_payoff(s):
    put_component = max(strike_price_put - s, 0) - premium_put
    call_component = 0 if s < strike_price_call else s - strike_price_call
    return (s - stock_price) + put_component - call_component + premium_call - premium_put

payoffs = [collar_payoff(s) for s in stock_prices]

plt.plot(stock_prices, payoffs)
plt.title('Collar Strategy Payoff')
plt.xlabel('Stock Price at Expiry')
plt.ylabel('Profit/Loss')
plt.grid(True)
plt.show()
```

### Exploring Advanced Options Risk Management Strategies

#### Iron Condors

```python
# Parameters for Iron Condor
lower_strike_put = 95
upper_strike_put = 100
lower_strike_call = 105
upper_strike_call = 110
premium_sold = 10  # Premium received for selling closer strikes
premium_bought = 4  # Premium paid for buying farther strikes

def iron_condor_payoff(s):
    payoff = min(s - lower_strike_put, 0) + max(s - upper_strike_put, 0)
    payoff += min(s - upper_strike_call, 0) + max(s - lower_strike_call, 0)
    return payoff - (premium_sold - premium_bought)

stock_prices = np.arange(90, 115, 1)
payoffs = [iron_condor_payoff(s) for s in stock_prices]

plt.plot(stock_prices, payoffs)
plt.title('Iron Condor Payoff')
plt.xlabel('Stock Price at Expiry')
plt.ylabel('Profit/Loss')
plt.grid(True)
plt.show()
```

#### Straddles

```python
# Parameters for Straddle
strike_price = 100
premium = 12  # Total premium paid for both the call and put

def straddle_payoff(s):
    return max(s - strike_price, 0) + max(strike_price - s, 0) - premium

payoffs = [straddle_payoff(s) for s in stock_prices]

plt.plot(stock_prices, payoffs)
plt.title('Long Straddle Payoff')
plt.xlabel('Stock Price at Expiry')
plt.ylabel('Profit/Loss')
plt.grid(True)
plt.show()
```

#### Strangles

```python
# Parameters for Strangle
lower_strike = 95
upper_strike = 105
premium = 10  # Total premium paid

def strangle_payoff(s):
    return max(s - upper_strike, 0) + max(lower_strike - s, 0) - premium

payoffs = [strangle_payoff(s) for s in stock_prices]

plt.plot(stock_prices, payoffs)
plt.title('Long Strangle Payoff')
plt.xlabel('Stock Price at Expiry')
plt.ylabel('Profit/Loss')
plt.grid(True)
plt.show()
```

#### Bull Spreads

```python
# Parameters for Bull Spread
lower_strike = 100
upper_strike = 105
premium_lower = 8
premium_upper = 3

def bull_spread_payoff(s):
    return max(s - lower_strike, 0) - premium_lower + premium_upper - max(s - upper_strike, 0)

payoffs = [bull_spread_payoff(s) for s in stock_prices]

plt.plot(stock_prices, payoffs)
plt.title('Bull Spread Payoff')
plt.xlabel('Stock Price at Expiry')
plt.ylabel('Profit/Loss')
plt.grid(True)
plt.show()
```

#### Bear Spreads

```python
# Parameters for Bear Spread
higher_strike = 105
lower_strike = 100
premium_higher = 9
premium_lower = 4

def bear_spread_payoff(s):
    return max(higher_strike - s, 0) - premium_higher + premium_lower - max(lower_strike - s, 0)

payoffs = [bear_spread_payoff(s) for s in stock_prices]

plt.plot(stock_prices, payoffs)
plt.title('Bear Spread Payoff')
plt.xlabel('Stock Price at Expiry')
plt.ylabel('Profit/Loss')
plt.grid(True)
plt.show()
```

#### Butterfly Spreads

```python
# Parameters for Butterfly Spread
lower_strike = 95
middle_strike = 100
upper_strike = 105
premium_lower = 3
premium_middle = 8
premium_upper = 2

def butterfly_payoff(s):
    return (max(s - lower_strike, 0) - 2 * max(s - middle_strike, 0) + max(s - upper_strike, 0) -
            premium_lower + 2 * premium_middle - premium_upper)

payoffs = [butterfly_payoff(s) for s in stock_prices]

plt.plot(stock_prices, payoffs)
plt.title('Butterfly Spread Payoff')
plt.xlabel('Stock Price at Expiry')
plt.ylabel('Profit/Loss')
plt.grid(True)
plt.show()
```

```python
np.random.seed(42)  # for reproducibility

# Mock returns and covariance matrix for three AAPL options
returns = np.array([0.08, 0.12, 0.07])
cov_matrix = np.array([[0.04, 0.02, 0.01], 
                       [0.02, 0.09, 0.03], 
                       [0.01, 0.03, 0.05]]) 

# Objective function to maximize the Sharpe Ratio
def negative_sharpe(weights): 
    port_return = np.dot(weights, returns)
    port_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    return -port_return / port_volatility

# Constraints and bounds for three options
cons = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1}) 
bounds = ((0, 1), (0, 1), (0, 1))
initial_weights = [1/3, 1/3, 1/3]  # Evenly distributed for the start

# Optimization
solution = minimize(negative_sharpe, initial_weights, method='SLSQP', bounds=bounds, constraints=cons)
optimal_weights = solution.x

# Monte Carlo Simulation for VaR
num_simulations = 10000
simulation_results = np.zeros(num_simulations)

for i in range(num_simulations):
    random_returns = np.random.multivariate_normal(mean_returns, cov_matrix)
    portfolio_return = np.dot(optimal_weights, random_returns)
    simulation_results[i] = portfolio_return

# Calculate VaR at 95% confidence level
VaR_95 = np.percentile(simulation_results, 5)
print(f"Value at Risk (VaR) at 95% Confidence Level: ${VaR_95:.2f}")
```
